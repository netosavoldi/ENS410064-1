```python
# Aula 05
## Análise de Séries Temporais
import pandas as pd
path = 'C:/Users/Windows 10/Documents/Python/ENS410064-1/scripts/aula_05.csv'
data = pd.read_csv(path)
data
```

```python
# Médias diárias
dailyAve = data.groupby(['year','month', 'day']).mean()
dailyAve = dailyAve.reset_index90
df = pd.DataFrame({'year': dailyAve['year'],
                   'month': dailyAve['month'],
                   'day': dailyAve['day']})
df = pd.to_datetime(df)
daiyAve['datetime']=df
dailyAve = dailyAve.set_index(['datetime'])


# plot dailyAve precipitacao
import matplotlib.pyplot as plt
dif,ax= plt.subplots()
ax.plot = dailyAve.index,dailyAve()
                         
```
